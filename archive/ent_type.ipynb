```python
import spacy
import os
nlp = spacy.load('en_core_web_sm')
```

```python
# Load entity translations
entities_dict = {}
with open( os.path.join('data/FB15K',  'entity2text.txt') ) as f:
    counter = 0
    for line in f.readlines():
        fields = line.split('\t')
        fields = [p.strip() for p in fields]
        # entities_dict[ fields[0] ] = fields[1]
        ent_txt = nlp(fields[1])
        if ent_txt.ents:
            for ent in ent_txt.ents:
                NER = str(spacy.explain(ent.label_))
        print(ent.text,":",NER)
        print('--')
        counter += 1
        if counter > 100:
            break
```

```python
import nltk
from nltk.chunk import ne_chunk
sent = 'Fast Food Nation'
for chunk in nltk.ne_chunk(nltk.pos_tag(nltk.word_tokenize(sent))):
     if hasattr(chunk, 'label'):
        print(chunk.label(), ' '.join(c[0] for c in chunk))
```
